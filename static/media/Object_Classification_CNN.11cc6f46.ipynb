# **What is convolution ?**

Convolution is a mathematical operation where we have an input I, and an argument, **kernel** K (or filter) to produce an output that expresses how the shape of one is modified by another.

**So, how do we implement that in CNN**?

The way we implement this is through Convolutional Layer. Convolutional layer is core building block of CNN, it helps with feature detection.

A kernel is a set of learnable filters and is small spatially compared to the image but extends through the full depth of the input image. A convolution layer transforms an input volume into an output volume of different size.

We have multiple kernels to help with things like edge detection, identifying different shapes, bends or different colors etc.

# **Pooling layers in CNN**

The pooling (POOL) layer reduces the height and width of the input. It helps reduce computation, as well as helps make feature detectors more invariant to its position in the input. Invariance here means that when we change the input by a small amount the pooled outputs do not change. This helps with detecting features that are common in the input like edges in an image or colors in an image.

The three types of pooling layers are:

* Max-pooling layer: slides an (𝑓,𝑓) window over the input and stores the 
max value of the window in the output.
* Min-pooling layer: slides an (𝑓,𝑓) window over the input and stores the 
min value of the window in the output.
* Average-pooling layer: slides an (𝑓,𝑓) window over the input and stores the average value of the window in the output.

# **Structure of a basic CNN**

So now we know that a convolutional network at the basic level consists of -

* Multiple convolutions performed in parallel — output is linear activation function
* Applying nonlinear function ReLU to the convolutional layers
* Use a pooling function like max pooling to summarize the statistics of nearby locations. This helps with “Translational Invariance”
* We flatten the max pooled output which are then inputs to a fully connected neural network

<center><img src="https://miro.medium.com/max/700/1*3Yf549FDfaqK9z3U1cx4MA.png" alt="example"></center>


Also in most of the cases we have to pad our images. The main benefits of **padding** are the following:

* It allows you to use a convolution layer without necessarily shrinking the height and width of the volumes. This is important for building deeper networks, since otherwise the height/width would shrink as you go to deeper layers. 

* It helps us keep more of the information at the border of an image. Without padding, very few values at the next layer would be affected by pixels as the edges of an image.
